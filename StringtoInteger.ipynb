```python
class Solution:
    def myAtoi(self, s: str):
        flag_positive = True
        s = s.strip()
        if s == "":
            return 0
        if s[0] in ("-", "+"):
            if s[0] == "-":
                flag_positive = False
            s = s[1:]

        final_sum = 0
        for i in range(len(s)):
            if not s[i].isdigit():
                break
            final_sum *= 10
            final_sum += int(s[i])

        if not flag_positive:
            final_sum *= -1
            if final_sum < -(2**31):
                return -(2**31)
        if final_sum > 2**31 - 1:
            return 2 ** 31 -1 
        return final_sum
```
